```python
# prompt: do the std min max and sum , average on perplexity for all the file paths, generate code for all languages

import pandas as pd

def calculate_perplexity_stats(file_path):
  """
  Calculates the sum, average, minimum, and maximum of perplexity scores in a given file.

  Args:
    file_path: The path to the CSV file containing perplexity scores.

  Returns:
    A dictionary containing the sum, average, minimum, and maximum of perplexity scores.
  """

  df = pd.read_csv(file_path)
  perplexities = df["perplexity"]
  return {
    "sum": perplexities.sum(),
    "average": perplexities.mean(),
    "min": perplexities.min(),
    "max": perplexities.max(),
  }

# Calculate perplexity stats for all languages
en_stats = calculate_perplexity_stats(en_file_path)
ar_stats = calculate_perplexity_stats(ar_file_path)
bn_stats = calculate_perplexity_stats(bn_file_path)
gd_stats = calculate_perplexity_stats(gd_file_path)
gn_stats = calculate_perplexity_stats(gn_file_path)
he_stats = calculate_perplexity_stats(he_file_path)
hi_stats = calculate_perplexity_stats(hi_file_path)
hmn_stats = calculate_perplexity_stats(hmn_file_path)
it_stats = calculate_perplexity_stats(it_file_path)
th_stats = calculate_perplexity_stats(th_file_path)
uk_stats = calculate_perplexity_stats(uk_file_path)
zh_stats = calculate_perplexity_stats(zh_file_path)
zu_stats = calculate_perplexity_stats(zu_file_path)

# Print the results
print("English perplexity stats:")
print(en_stats)

print("Arabic perplexity stats:")
print(ar_stats)

print("Bengali perplexity stats:")
print(bn_stats)

print("Scottish Gaelic perplexity stats:")
print(gd_stats)

print("Guarani perplexity stats:")
print(gn_stats)

print("Hebrew perplexity stats:")
print(he_stats)

print("Hindi perplexity stats:")
print(hi_stats)

print("Hmong perplexity stats:")
print(hmn_stats)

print("Italian perplexity stats:")
print(it_stats)

print("Thai perplexity stats:")
print(th_stats)

print("Ukrainian perplexity stats:")
print(uk_stats)

print("Chinese perplexity stats:")
print(zh_stats)

print("Zulu perplexity stats:")
print(zu_stats)
```

```python
# prompt: store all of these stats in a dataframe and use "language name" as a column name where the row denotes the language

import pandas as pd

# Create a list of dictionaries containing the perplexity stats for each language.
language_stats = [
  {"language": "English", **en_stats},
  {"language": "Arabic", **ar_stats},
  {"language": "Bengali", **bn_stats},
  {"language": "Scottish Gaelic", **gd_stats},
  {"language": "Guarani", **gn_stats},
  {"language": "Hebrew", **he_stats},
  {"language": "Hindi", **hi_stats},
  {"language": "Hmong", **hmn_stats},
  {"language": "Italian", **it_stats},
  {"language": "Thai", **th_stats},
  {"language": "Ukrainian", **uk_stats},
  {"language": "Chinese", **zh_stats},
  {"language": "Zulu", **zu_stats},
]

# Create a Pandas DataFrame from the list of dictionaries.
df = pd.DataFrame(language_stats)

# Print the DataFrame.
print(df)
```
